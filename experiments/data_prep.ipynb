```python
import numpy as np
import os
from data_utils import Utils
import panads as pd
```

```python
channels = [["FC1", "FC2"],
            ["FC3", "FC4"],
            ["FC5", "FC6"],
            ["C5", "C6"],
            ["C3", "C4"],
            ["C1", "C2"],
            ["CP1", "CP2"],
            ["CP3", "CP4"],
            ["CP5", "CP6"]]
```

```python
exclude = [38, 88, 89, 92, 100, 104] # corrupted experiments to exculde
subjects = [n for n in np.arange(1, 110) if n not in exclude]
runs = [4, 6, 8, 10, 12, 14]
```

```python
data_path = '../physionet.org/files/eegmmidb/1.0.0'
prep_data_path = 'data/'
if not os.path.isdir(prep_data_path):
    os.mkdir(prep_data_path)
```

```python
for couple in channels:
    save_path = os.path.join(prep_data_path, couple[0] + couple[1])
    if not os.path.isdir(save_path):
        os.mkdir(save_path)
    for sub in subjects:
        x, y = Utils.epoch(Utils.select_channels
            (Utils.eeg_settings(Utils.del_annotations(Utils.concatenate_runs(
            Utils.load_data(subjects=[sub], runs=runs, data_path=data_path)))), couple),
            exclude_base=False)

        np.save(os.path.join(save_path, "x_sub_" + str(sub)), x, allow_pickle=True)
        np.save(os.path.join(save_path, "y_sub_" + str(sub)), y, allow_pickle=True)
```
